## Temporal Feature Extraction from timestamp

**Source Column:** timestamp

**Feature to Create**
- **hour**: Hour of Interaction (0-23)
- **day_of_week**: Day of week (0 - 6)
- **day_of_month**: Day of Month (1 - 31)
- **month**: Month of Year (1-12)
- **year**: Year
- **is_weekend**: Binary flag (1 for Sat/Sun) 0 otherwise
- **day_period**: Time period category (morning, afternoon, evening, night)

**Purpose:** Identify temporal patterns in user interactions

```python
#converting timestamp to datetime
df1['timestamp'] = pd.to_datetime(df1['timestamp'], unit='s', errors='coerce')

#extracting temporal features
df1['hour'] = df1['timestamp'].dt.hour
df1['day_of_week'] = df1['timestamp'].dt.dayofweek
df1['day_of_month'] = df1['timestamp'].dt.day
df1['month'] = df1['timestamp'].dt.month
df1['year'] = df1['timestamp'].dt.year
df1['is_weekend'] = df1['day_of_week'].isin([5, 6]).astype(int)

#creating day period feature
def assign_day_period(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df1['day_period'] = df1['hour'].apply(assign_day_period)

print("Extracted temporal features and created 'day_period' feature.")
print(df1[['timestamp', 'hour', 'day_of_week', 'day_of_month', 'month', 'year', 'is_weekend', 'day_period']].head(10))
print(f"Unique values in 'day_period': {df1['day_period'].value_counts()}")
```

## Outlier Detection (IQR Method)

**Columns to Check for Outliers:**
- **price**: Product price
- **discountPercentage**: Discount percentage
- **stock**: Inventory levels
- **weight**: Product weight
- **user_rating**: User ratings (should be 0-5)
- **product_rating**: Product ratings (should be 0-5)

**Method:** Interquartile Range (IQR)
- Calculate Q1 (25th percentile) and Q3 (75th percentile)
- IQR = Q3 - Q1
- Lower bound = Q1 - 1.5 × IQR
- Upper bound = Q3 + 1.5 × IQR

**Action:** Flag outliers but keep them for analysis (don't delete)

```python
## Outliers Detection (IQR Method)
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = (df[column] < lower_bound) | (df[column] > upper_bound)
    return outliers, lower_bound, upper_bound
```

```python
numerical_columns = ['price', 'discountPercentage', 'stock', 'weight', 'user_rating', 'product_rating']

#Detection of outliers for each numerical column
outlier_summary = {}
for col in numerical_columns:
    outliers, lower_bound, upper_bound = detect_outliers_iqr(df1, col)
    cnt = outliers.sum()
    outlier_summary[col] = {
        'count': cnt,
        'outlier_count': outliers.sum(),
        'percentage': (cnt / len(df1)) * 100,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound
    }

    print(f"\n{col} Outliers:"
          f"\nLower Bound: {lower_bound:.2f}"
          f"\nUpper Bound: {upper_bound:.2f}"
          f"\nOutliers Count: {cnt:,} ({(cnt / len(df1)) * 100:.2f}%)")
    

# Creating outlier flag column
df1['has_outlier'] = False
for col in numerical_columns:
    outliers, lower_bound, upper_bound = detect_outliers_iqr(df1, col)
    df1.loc[outliers, 'has_outlier'] = True

print(f"\nTotal records with outliers in any numerical column: {df1['has_outlier'].sum():,} ({(df1['has_outlier'].mean() * 100):.2f}%)")
```

## Data Normalization

**Purpose:** Scale numerical features to a standard range [0, 1]

**Columns to Normalize:**
- price
- discountPercentage
- stock
- weight
- user_rating
- product_rating

**Method:** MinMax Scaling
- Formula: X_normalized = (X - X_min) / (X_max - X_min)
- Result: All values fall between 0 and 1
- Advantage: Preserves the shape of original distribution
- Use Case: Machine learning models sensitive to feature scaling

**Why:** Different features have different ranges (price 0-5000, ratings 0-5). Normalization ensures equal importance in downstream analysis.

```python
from sklearn.preprocessing import MinMaxScaler

# Columns to normalize
numerical_cols_to_normalize = ['price', 'discountPercentage', 'stock', 'weight', 'user_rating', 'product_rating']

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply normalization
df1[['price_normalized', 'discount_normalized', 'stock_normalized', 
     'weight_normalized', 'user_rating_normalized', 'product_rating_normalized']] = \
    scaler.fit_transform(df1[numerical_cols_to_normalize])

print("Data Normalization Completed:")
print("=" * 70)
print("\nNormalized Columns Statistics:")
normalized_cols = ['price_normalized', 'discount_normalized', 'stock_normalized', 
                   'weight_normalized', 'user_rating_normalized', 'product_rating_normalized']
print(df1[normalized_cols].describe())

print("\nSample comparison (Original vs Normalized):")
comparison_df = pd.DataFrame({
    'price_original': df1['price'].head(),
    'price_normalized': df1['price_normalized'].head(),
    'rating_original': df1['user_rating'].head(),
    'rating_normalized': df1['user_rating_normalized'].head()
})
print(comparison_df)
```

## Data Standardization (Z-Score Scaling)

**Purpose:** Transform features to have mean = 0 and standard deviation = 1

**Columns to Standardize:**
- price
- discountPercentage
- stock
- weight

**Method:** StandardScaler (Z-Score)
- Formula: X_standardized = (X - mean) / std_dev
- Result: Bell-curve centered at 0
- Advantage: Better for algorithms assuming normally distributed data
- Use Case: Linear/Logistic Regression, SVM, Neural Networks

**When to Use:** When data follows normal distribution or when comparing across different scales

```python
from sklearn.preprocessing import StandardScaler

# Columns to standardize
numerical_cols_to_standardize = ['price', 'discountPercentage', 'stock', 'weight']

# Initialize StandardScaler
std_scaler = StandardScaler()

# Apply standardization
df1[['price_standardized', 'discount_standardized', 'stock_standardized', 
     'weight_standardized']] = \
    std_scaler.fit_transform(df1[numerical_cols_to_standardize])

print("Data Standardization (Z-Score) Completed:")
print("=" * 70)
print("\nStandardized Columns Statistics (should have mean≈0, std≈1):")
standardized_cols = ['price_standardized', 'discount_standardized', 
                     'stock_standardized', 'weight_standardized']
stats = df1[standardized_cols].describe()
print(stats)

print("\nVerification (Mean should be ≈ 0, Std should be ≈ 1):")
for col in standardized_cols:
    print(f"{col}: Mean = {df1[col].mean():.6f}, Std = {df1[col].std():.6f}")
```

## Data Binning (Discretization)

**Purpose:** Convert continuous numerical values into discrete categorical bins

**Columns to Bin:**
1. **price** → Price ranges (Budget, Mid-range, Premium, Luxury)
2. **discountPercentage** → Discount levels (No Discount, Low, Medium, High)
3. **product_rating** → Rating categories (Poor, Below Average, Average, Good, Excellent)
4. **stock** → Inventory levels (Out of Stock, Low, Medium, High)

**Method:** Equal-width binning based on domain knowledge

**Benefits:**
- Simplifies data analysis
- Reduces impact of outliers
- Creates interpretable categories
- Useful for visualization and business rules

```python
# 1. Price Binning
price_max = df1['price'].max()
price_bins = [0, 50, 200, 500, price_max + 1]
price_labels = ['Budget', 'Mid-range', 'Premium', 'Luxury']
df1['price_category'] = pd.cut(df1['price'], bins=price_bins, labels=price_labels, include_lowest=True)

# 2. Discount Percentage Binning
discount_bins = [-0.1, 0, 20, 50, 101]
discount_labels = ['No Discount', 'Low', 'Medium', 'High']
df1['discount_category'] = pd.cut(df1['discountPercentage'], bins=discount_bins, 
                                   labels=discount_labels, include_lowest=True)

# 3. Product Rating Binning
rating_bins = [0, 1, 2, 3, 4, 5.1]
rating_labels = ['Poor', 'Below Average', 'Average', 'Good', 'Excellent']
df1['product_rating_category'] = pd.cut(df1['product_rating'], bins=rating_bins, 
                                         labels=rating_labels, include_lowest=True)

# 4. Stock Level Binning (FIXED - Dynamic bins)
stock_min = df1['stock'].min()
stock_max = df1['stock'].max()
stock_bins = [stock_min - 1, 10, 30, 60, stock_max + 1]  # Fixed: strictly increasing values
stock_labels = ['Out of Stock', 'Low', 'Medium', 'High']
df1['stock_category'] = pd.cut(df1['stock'], bins=stock_bins, labels=stock_labels, include_lowest=True)

print("Data Binning Completed:")
print("=" * 70)

print("\n1. Price Categories:")
print(f"   Bins: {price_bins}")
print(df1['price_category'].value_counts().sort_index())

print("\n2. Discount Categories:")
print(f"   Bins: {discount_bins}")
print(df1['discount_category'].value_counts().sort_index())

print("\n3. Product Rating Categories:")
print(f"   Bins: {rating_bins}")
print(df1['product_rating_category'].value_counts().sort_index())

print("\n4. Stock Categories:")
print(f"   Bins: {stock_bins}")
print(df1['stock_category'].value_counts().sort_index())

print("\nSample binned data:")
print(df1[['price', 'price_category', 'discountPercentage', 'discount_category', 
           'product_rating', 'product_rating_category', 'stock', 'stock_category']].head(10))
```

## Data Wrangling

**Purpose:** Clean, transform, and restructure data for analysis

**Operations:**
1. Handle inconsistent category values (standardize text)
2. Remove special characters from text fields
3. Handle unexpected data types
4. Validate and fix data format issues
5. Remove or encode categorical outliers
6. Handle zero/negative values in specific columns

**Why Important:** Ensures consistency and quality for downstream analysis and modeling

```python
# 1. Standardize categorical columns (lowercase, strip whitespace)
categorical_columns_to_clean = ['category', 'interaction_type', 'availabilityStatus']

for col in categorical_columns_to_clean:
    if col in df1.columns:
        df1[col] = df1[col].str.lower().str.strip()
        print(f"{col} unique values: {df1[col].nunique()}")
        print(df1[col].value_counts().head())
        print()

# 2. Handle interaction_type - ensure only valid types
valid_interactions = ['purchase', 'view', 'cart', 'add', 'remove', 'wishlist']
print(f"Unique interaction types in data:")
print(df1['interaction_type'].unique())

# 2. Fix availability status - standardize values
print(f"\nAvailability Status - Before wrangling:")
print(df1['availabilityStatus'].value_counts())

# Map similar values
availability_mapping = {
    'in stock': 'in stock',
    'out of stock': 'out of stock',
    'limited stock': 'limited stock',
    'back order': 'back order',
    'discontinued': 'discontinued'
}

df1['availabilityStatus'] = df1['availabilityStatus'].map(
    lambda x: availability_mapping.get(x, x) if x in availability_mapping.values() else 'unknown'
)

print(f"\nAvailability Status - After wrangling:")
print(df1['availabilityStatus'].value_counts())

# 3. Ensure numerical columns have no unexpected strings
# Convert any non-numeric values to NaN, then fill with median/default
df1['stock'] = pd.to_numeric(df1['stock'], errors='coerce').fillna(df1['stock'].median())
df1['weight'] = pd.to_numeric(df1['weight'], errors='coerce').fillna(df1['weight'].median())

# 4. Fix negative values where they shouldn't exist
df1['stock'] = df1['stock'].clip(lower=0)  # Stock can't be negative
df1['weight'] = df1['weight'].clip(lower=0.1)  # Weight should be positive

# 5. Ensure price is always positive
df1['price'] = df1['price'].clip(lower=0.01)  # Minimum price

# 6. Discount percentage must be 0-100
df1['discountPercentage'] = df1['discountPercentage'].clip(0, 100)

# 7. Ratings must be 0-5
df1['user_rating'] = df1['user_rating'].clip(0, 5)
df1['product_rating'] = df1['product_rating'].clip(0, 5)

print("\nData Wrangling Completed:")
print("=" * 70)
print(" Fixed availability status values")
print(" Converted non-numeric values")
print(" Standardized categorical columns")
print(" Removed negative values from non-negative columns")
print(" Constrained values to valid ranges")

print("\nData ranges after wrangling:")
print(f"Price range: {df1['price'].min():.2f} - {df1['price'].max():.2f}")
print(f"Discount range: {df1['discountPercentage'].min():.2f}% - {df1['discountPercentage'].max():.2f}%")
print(f"Stock range: {df1['stock'].min():.0f} - {df1['stock'].max():.0f}")
print(f"User rating range: {df1['user_rating'].min():.1f} - {df1['user_rating'].max():.1f}")
print(f"Product rating range: {df1['product_rating'].min():.1f} - {df1['product_rating'].max():.1f}")
```
